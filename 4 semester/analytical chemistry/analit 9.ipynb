```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

tm = 1.285
L = 1

df = pd.DataFrame({})

df["names"] = ["$\ce{F^-}$", "$\ce{Cl^-}$", "$\ce{NO2^-}$", "$\ce{Br^-}$", 
               "$\ce{NO3^-}$", "$\ce{HPO4^2-}$", "$\ce{SO4^2-}$"]
df["tR1"]   = [2.3,3.853,5.873, 7.493, 11.157, 14.087, 18.227]
df["tR2"]   = df.tR1 - tm
df["k"] = (df.tR1 - tm) / tm
df["W"] = [0.46, 0.71, 0.93, 1.76, 1.59, 1.97, 2.89] 
df["N"] = 16 * (df.tR2 / df.W)**2
#df["H"] = L*1000 / df.N 


display(df)
from tabulate import tabulate

def format_dataframe(df):
    
    def auto_format(number):
        if isinstance(number, (int, float)):
            if abs(number) >= 1000 or (abs(number) < 0.001 and number != 0):
                return f"{number:.3e}"  # Экспоненциальный формат
            else:
                return f"{number:.3f}"  # Обычный формат
        return str(number)  # Если это не число, оставляем как есть
    
    return df.applymap(auto_format)

print(tabulate(format_dataframe(df),
              tablefmt = "github",
              headers = ["Ионы", "$t_{R_1}$","$t_{R_2}$","k", "W", "N"]))
```
