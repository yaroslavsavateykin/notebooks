```python
def format_dataframe(df):
    
    def auto_format(number):
        if isinstance(number, (int, float)):
            if abs(number) >= 1000 or (abs(number) < 0.001 and number != 0):
                return f"{number:.3e}"  # Экспоненциальный формат
            else:
                return f"{number:.3f}"  # Обычный формат
        return str(number)  # Если это не число, оставляем как есть
    
    return df.applymap(auto_format)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tabulate import tabulate
from IPython.display import display

tm = 0.196 
F = 30

df1 = pd.DataFrame({})

df1["Названия"] =  ["н-гексан", "н-гептан", "н-октан", "н-декан"]
df1["tR1"] = [0.458, 0.748, 1.396, 5.896]
df1["tR2"] = df1.tR1 - tm 
df1["VR1"]  = F * df1.tR1
df1["VR2"] = F * df1.tR2
df1["k"] = (df1.tR1 - tm) / tm

#print(tabulate(df1, tablefmt = "github"), floatfmt = ".2f")

display(df1)

print(tabulate(format_dataframe(df1),
              tablefmt = "github",
              headers = ["Углеводороды", "$t_R$, мин", "$t_R'$, мин",
                         "$V_R$, мл", "$V_R'$, мл", "$k'$"]))
```
