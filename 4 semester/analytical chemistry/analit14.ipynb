```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit






# Исходные данные
c_HA = np.array([6e-6, 9e-6, 1.2e-5, 1.5e-5, 1.8e-5])
A_HA_lambda1 = np.array([0.2436, 0.3674, 0.4682, 0.5891, 0.6922])
A_HA_lambda2 = np.array([0.0584, 0.0733, 0.0898, 0.1074, 0.1279])

c_A_minus = np.array([9e-6, 1.2e-5, 1.5e-5, 1.8e-5, 2.1e-5, 2.4e-5])
A_A_minus_lambda1 = np.array([0.0636, 0.0986, 0.0964, 0.1095, 0.1258, 0.1394])
A_A_minus_lambda2 = np.array([0.2039, 0.2615, 0.3236, 0.3791, 0.4398, 0.4982])

# Линейная модель
def linear(x, a):
    return a * x

# Функция для расчета R²
def manual_r2(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred)**2)
    ss_tot = np.sum((y_true - np.mean(y_true))**2)
    return 1 - (ss_res / ss_tot)

# Построение графиков
############################################################
# График 1: HA при lambda = 530 nm
fig1, ax1 = plt.subplots(figsize=(8, 5))
popt, _ = curve_fit(linear, c_HA, A_HA_lambda1)
y_fit = linear(c_HA, *popt)
r2 = manual_r2(A_HA_lambda1, y_fit)

ax1.scatter(c_HA*1e6, A_HA_lambda1, s=80, label='Экспериментальные точки')
ax1.plot(c_HA*1e6, y_fit, 'r-', label=f'Аппроксимация: $A = {popt[0]:.1f}c$ ($R^2 = {r2:.4f}$)')
ax1.set_title('Кислотная форма HA ($\\lambda = 530$ нм)', fontsize=14)
ax1.set_xlabel('Концентрация, $\\times 10^{-6}$ М', fontsize=12)
ax1.set_ylabel('Оптическая плотность', fontsize=12)
ax1.legend()
plt.tight_layout()
plt.savefig('HA_lambda1.png', dpi=300)
plt.close()

############################################################
# График 2: HA при lambda = 422 nm
fig2, ax2 = plt.subplots(figsize=(8, 5))
popt, _ = curve_fit(linear, c_HA, A_HA_lambda2)
y_fit = linear(c_HA, *popt)
r2 = manual_r2(A_HA_lambda2, y_fit)

ax2.scatter(c_HA*1e6, A_HA_lambda2, s=80, label='Экспериментальные точки')
ax2.plot(c_HA*1e6, y_fit, 'r-', label=f'Аппроксимация: $A = {popt[0]:.1f}c$ ($R^2 = {r2:.4f}$)')
ax2.set_title('Кислотная форма HA ($\\lambda = 422$ нм)', fontsize=14)
ax2.set_xlabel('Концентрация, $\\times 10^{-6}$ М', fontsize=12)
ax2.set_ylabel('Оптическая плотность', fontsize=12)
ax2.legend()
plt.tight_layout()
plt.savefig('HA_lambda2.png', dpi=300)
plt.close()

############################################################
# График 3: A⁻ при lambda = 530 nm
fig3, ax3 = plt.subplots(figsize=(8, 5))
popt, _ = curve_fit(linear, c_A_minus, A_A_minus_lambda1)
y_fit = linear(c_A_minus, *popt)
r2 = manual_r2(A_A_minus_lambda1, y_fit)

ax3.scatter(c_A_minus*1e6, A_A_minus_lambda1, s=80, label='Экспериментальные точки')
ax3.plot(c_A_minus*1e6, y_fit, 'r-', label=f'Аппроксимация: $A = {popt[0]:.1f}c$ ($R^2 = {r2:.4f}$)')
ax3.set_title('Основная форма A⁻ ($\\lambda = 530$ нм)', fontsize=14)
ax3.set_xlabel('Концентрация, $\\times 10^{-6}$ М', fontsize=12)
ax3.set_ylabel('Оптическая плотность', fontsize=12)
ax3.legend()
plt.tight_layout()
plt.savefig('Aminus_lambda1.png', dpi=300)
plt.close()

############################################################
# График 4: A⁻ при lambda = 422 nm
fig4, ax4 = plt.subplots(figsize=(8, 5))
popt, _ = curve_fit(linear, c_A_minus, A_A_minus_lambda2)
y_fit = linear(c_A_minus, *popt)
r2 = manual_r2(A_A_minus_lambda2, y_fit)

ax4.scatter(c_A_minus*1e6, A_A_minus_lambda2, s=80, label='Экспериментальные точки')
ax4.plot(c_A_minus*1e6, y_fit, 'r-', label=f'Аппроксимация: $A = {popt[0]:.1f}c$ ($R^2 = {r2:.4f}$)')
ax4.set_title('Основная форма A⁻ ($\\lambda = 422$ нм)', fontsize=14)
ax4.set_xlabel('Концентрация, $\\times 10^{-6}$ М', fontsize=12)
ax4.set_ylabel('Оптическая плотность', fontsize=12)
ax4.legend()
plt.tight_layout()
plt.savefig('Aminus_lambda2.png', dpi=300)
plt.close()
```
