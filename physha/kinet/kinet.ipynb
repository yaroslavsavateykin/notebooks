```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import os

SAVE_DIR = "/home/yaroslav/vzlet_vault/Практикум/kinet/"

def savefig(name, path="", fig=None, bbox_inches=None, **kwargs):
    os.makedirs(SAVE_DIR, exist_ok=True)
    if bbox_inches is None:
        bbox_inches = "tight"
    if fig is None:
        fig = plt.gcf()
    fig.savefig(os.path.join(SAVE_DIR, name), bbox_inches=bbox_inches, **kwargs)
    fig.savefig(os.path.join(".", name), bbox_inches=bbox_inches, **kwargs)


A = 1.0

param_sets = [
    ("слабый автокатализ", "slabyi_autokat", 0.5, 0.3, 0.5),
    ("сильный автокатализ", "silnyi_autokat", 1.0, 1.0, 0.3),
    ("почти колебания", "pochti_kolebaniya", 1.2, 1.5, 0.2),
]

def rhs(t, y, k1, k2, k3):
    X, Y = y
    return [k1*A - k2*X*Y, k2*X*Y - k3*Y]

t = np.linspace(0, 40, 4000)

for rus, eng, k1, k2, k3 in param_sets:
    sol = solve_ivp(rhs, [0, 40], [0.1, 0.1], args=(k1, k2, k3), t_eval=t)

    fig, ax = plt.subplots(figsize=(10,5), dpi=150)
    ax.grid()
    ax.plot(t, sol.y[0], label="X")
    ax.plot(t, sol.y[1], label="Y")

    ax.set_title(f"Lotka: {rus};  ($k_1$={k1}, $k_2$={k2}, $k_3$={k3})")
    ax.set_xlabel("t")
    ax.legend()

    savefig(f"lotka_{eng}.png", fig=fig)
    plt.close(fig)
```
