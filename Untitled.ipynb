```python
import pandas as pd
import numpy as np
from unc_tools import UncRegression
from uncertainties.umath import exp

data = {
    'temp': [529.5, 538.5, 550.5, 560.5, 569.5],
    't_05': [66.5, 44.3, 29.6, 19.3, 14.6]
}

df = pd.DataFrame(data)

C0 = 1.8e16 / 6.02e23 * 1000
print(C0)

df["k"] = 1 / (df.t_05 * C0)
df.temp += 273.15


reg = UncRegression(1 / df.temp, np.log(df.k))
reg.latex_style()
reg.plot()

ear, lna = reg.coefs

print(-ear * 8.314)
print(f"{exp(lna):5e}")
```
